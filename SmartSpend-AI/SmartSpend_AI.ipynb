```python
# Select required columns
df = df[[
    "trans_date_trans_time",
    "merchant",
    "amt",
    "category"
]]
```

```python
# Rename columns
df = df.rename(columns={
    "trans_date_trans_time": "Date",
    "merchant": "Description",
    "amt": "Amount",
    "category": "Category"
})
```

```python
# Convert Date
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Handle missing values
df["Description"].fillna("Unknown", inplace=True)
df["Category"].fillna("Unknown", inplace=True)

# Drop rows where Date or Amount is missing
df.dropna(subset=["Date", "Amount"], inplace=True)
```

```python
print(df.isnull().sum())
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# Convert date
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
#NLP: TRANSACTION CATEGORIZATION
X_text = df["Description"]
y = df["Category"]

tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=3000
)

X_tfidf = tfidf.fit_transform(X_text)

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)

clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Example prediction
sample_txn = ["AMAZON ONLINE SHOPPING"]
sample_vec = tfidf.transform(sample_txn)
print("Predicted Category:", clf.predict(sample_vec)[0])
```

```python
#3. #ANOMALY DETECTION (UNUSUAL SPENDING)
# ------------------------------
print("\n--- Unusual Spending Detection ---")

iso = IsolationForest(
    n_estimators=100,
    contamination=0.05,
    random_state=42
)

df["Anomaly"] = iso.fit_predict(df[["Amount"]])
df["Anomaly"] = df["Anomaly"].map({1: "Normal", -1: "Unusual"})

print("\nTop Unusual Transactions:")
print(df[df["Anomaly"] == "Unusual"].head())
```

```python
print("\n--- Monthly Budget Prediction ---")

df["Month"] = df["Date"].dt.to_period("M")
monthly = df.groupby("Month")["Amount"].sum().reset_index()

monthly["Month_Index"] = np.arange(len(monthly))

X = monthly[["Month_Index"]]
y = monthly["Amount"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lr = LinearRegression()
lr.fit(X_scaled, y)

next_month = pd.DataFrame(
    [[monthly["Month_Index"].max() + 1]],
    columns=["Month_Index"]
)
next_month_scaled = scaler.transform(next_month)
predicted_budget = lr.predict(next_month_scaled)
predicted_budget = lr.predict(next_month)

print("💰 Predicted next month spending: ₹", round(predicted_budget[0], 2))
```

```python
# 5. VISUALIZATION
# ------------------------------
plt.figure()
plt.plot(monthly["Month"].astype(str), monthly["Amount"], marker="o")
plt.xticks(rotation=45)
plt.title("Monthly Spending Trend")
plt.xlabel("Month")
plt.ylabel("Total Amount")
plt.tight_layout()
plt.show()

print("\n✅ SmartSpend AI executed successfully")
```
